In almost ML projects the six steps shall follow:


1- Read and Load data,


2- Preprocessing the dataset--> Data Cleaning, 


3- Split dataset--> split to X(features) and y(labels), then split to train and test set,


4- Call modules and create an instance from called class,


5- Fit model and train on the dataset, 


6- Assess model performance and visual outcomes. 



To create an ML project: a library such as: sklearn--> get access to certain Module, i.e. linear_model -->from Module
we call a certain Class, i.e. LinearRegression,-->create an instance from Class

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
data = pd.read_csv('Salary_Data.csv')
```

```python
# play around data to get feeling of it
# df.keys(); is almsot same as key :value. keys are header and values are the value under each key in a corresponding column. 
""" by df.keys() we get the headers == keys and to access to the values under each key == corresponding values in a column; df['key'].values"""
data.keys() # Index(['YearsExperience', 'Salary'], dtype='object')
#info() method returns great information about dataset. 
data.info() 
# corr invesigate the relation between columns, Note: corr ignores the non-numeric values. 
"""the correlaion is a diagonal symmetric square marix. The leading diagonal is 1, due to corr(X,X)=1 = var(X). 
"""
data.corr()
```

```python
# Split dataset to X & y
X = data.iloc[:, :1].values
y = data.iloc[:, -1].values
```

```python
# Split data set--> train and test set
from sklearn.model_selection import train_test_split
# Above function returns tuples, thus unpack them, 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

```python
# Call module,model and class
from sklearn.linear_model import LinearRegression

# Create Instance
regressor = LinearRegression()

# Train 
regressor.fit(X_train, y_train)

# Predict 
predict = regressor.predict(X_test)
```

```python
# plot data and visualize outcome--> for training sets
plt.scatter(X_train, y_train, c = 'g')
plt.plot(X_train, regressor.predict(X_train), color = 'k')
plt.xlabel('years of Experience')
plt.ylabel('Salary')
plt.title('Salary & years of Experience')
plt.show()
```

```python
# plot data and visualize outcome--> for test sets
plt.scatter(X_test, y_test, color = 'navy')
# the line is for predicted salary in a same line. However, by using x_test the line
# is same, coz the equation for line is unique.
plt.plot(X_train, regressor.predict(X_train), color = 'darkviolet')
plt.xlabel('years of Experience')
plt.ylabel('Salary')
plt.title('Salary & years of Experience')
plt.show()
```
